```python
import pyodbc
import pandas as pd
```

```python
NursingHomesFile="../DataFiles/COVID-19_Nursing_Home_Dataset.csv"
```

```python
NursingHomesData = pd.read_csv(NursingHomesFile, iterator=True, chunksize=1000)
df = pd.concat(chunk[(chunk['Submitted Data'] =='Y') & (chunk['Passed Quality Assurance Check']=='Y')] for chunk in NursingHomesData)
df.head(10)
```

```python
del df['Federal Provider Number']
del df['Provider Name']
del df['Provider Address']
del df['Provider City']
del df['Provider Zip Code']
del df['Submitted Data']
del df['Passed Quality Assurance Check']
del df['Residents Weekly Suspected COVID-19']
del df['Residents Total Suspected COVID-19']
del df['Residents Weekly All Deaths']
del df['Residents Total All Deaths']
del df['Resident Access to Testing in Facility']
del df['Laboratory Type Is State Health Dept']
del df['Laboratory Type Is Private Lab']
del df['Laboratory Type Is Other']
del df['Staff Weekly Suspected COVID-19']
del df['Staff Total Suspected COVID-19']
del df['Shortage of Nursing Staff']
del df['Shortage of Clinical Staff']
del df['Shortage of Aides']
del df['Shortage of Other Staff']
del df['Any Current Supply of N95 Masks']
del df['One-Week Supply of N95 Masks']
del df['Any Current Supply of Surgical Masks']
del df['One-Week Supply of Surgical Masks']
del df['Any Current Supply of Eye Protection']
del df['One-Week Supply of Eye Protection']
del df['Any Current Supply of Gowns']
del df['One-Week Supply of Gowns']
del df['Any Current Supply of Gloves']
del df['One-Week Supply of Gloves']
del df['Any Current Supply of Hand Sanitizer']
del df['One-Week Supply of Hand Sanitizer']
del df['Ventilator Dependent Unit']
del df['Number of Ventilators in Facility']
del df['Number of Ventilators in Use for COVID-19']
del df['Any Current Supply of Ventilator Supplies']
del df['One-Week Supply of Ventilator Supplies']
del df['Total Resident Confirmed COVID-19 Cases Per 1,000 Residents']
del df['Total Resident COVID-19 Deaths Per 1,000 Residents']
del df['Total Residents COVID-19 Deaths as a Percentage of Confirmed COVID-19 Cases']
del df['Three or More Confirmed and Suspected COVID-19 Cases This Week']
del df['Initial Confirmed COVID-19 Case This Week']
del df['Geolocation']

del df['Number of All Beds']
del df['Total Number of Occupied Beds']
del df['Residents Weekly Admissions COVID-19']
del df['Residents Total Admissions COVID-19']
df.head(10)
```

```python
df['Week Ending']=pd.to_datetime(df['Week Ending'])
df=df.assign(WeekNo=lambda x: (x['Week Ending'].dt.isocalendar().week))
df.head(10)
```

```python
df=df[['Week Ending','WeekNo','Provider State','County',
       'Residents Weekly Confirmed COVID-19',
       'Residents Total Confirmed COVID-19',
       'Residents Weekly COVID-19 Deaths',
       'Residents Total COVID-19 Deaths',
       'Staff Weekly Confirmed COVID-19',
       'Staff Total Confirmed COVID-19',
       'Staff Weekly COVID-19 Deaths',
       'Staff Total COVID-19 Deaths']]
df.head(10)
```

```python
dfClean=df.where(pd.notnull(df),None)
dfClean.head(10)
```

```python
dfSum=dfClean.groupby(by=['Week Ending','WeekNo','Provider State','County'], as_index=False).agg({
            'Residents Weekly Confirmed COVID-19':'sum',
            'Residents Total Confirmed COVID-19':'sum',
            'Residents Weekly COVID-19 Deaths':'sum',
            'Residents Total COVID-19 Deaths':'sum',
            'Staff Weekly Confirmed COVID-19':'sum',
            'Staff Total Confirmed COVID-19':'sum',
            'Staff Weekly COVID-19 Deaths':'sum',
            'Staff Total COVID-19 Deaths':'sum'})
dfSum.head(10)
```

```python
NursingHomesList=dfSum.values.tolist()
```
